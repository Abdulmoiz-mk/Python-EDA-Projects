```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/covid19_Confirmed_dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['Lat', 'Long'])
```

```python
df.head()
```

```python
aggregation = df.groupby('Country/Region').sum()
aggregation.head()
```

```python
df.shape
```

```python
#visualize the data for china, italy and spain
```

```python
aggregation.loc['China'].plot()
aggregation.loc['Italy'].plot()
aggregation.loc['Spain'].plot()
plt.legend()
plt.show()
```

```python
#Maximum infection rate
```

```python
#Max infection for china
aggregation.loc['China'].diff().max()
```

```python
aggregation.loc['Italy'].diff().max()
```

```python
aggregation.loc['Spain'].diff().max()
```

```python
countries = list(aggregation.index)
max_infection_rate = []

for c in countries:
  max_infection_rate.append(aggregation.loc[c].diff().max())
aggregation['max_infection_rate'] = max_infection_rate
```

```python
aggregation.head()
```

```python
#get only required dataset

data = pd.DataFrame(aggregation['max_infection_rate'])
data.head()
```

```python
happiness = pd.read_csv("/content/worldwide_happiness_report.csv")
happiness.head()
```

```python
cols_to_drop = ['Overall rank', 'Score', "Generosity", "Perceptions of corruption"]
happiness.drop(columns = cols_to_drop, inplace=True)
happiness.head()
```

```python
happiness.head()
```

```python
happiness.set_index('Country or region', inplace = True)
happiness.head()
```

```python
happiness.head()
```

```python
data.head()
```

```python
final_data = data.join(happiness, how='inner')
final_data.head()
```

```python
final_data.shape
```

```python
cm = final_data.corr()
sns.heatmap(cm, annot=True)
```

```python
x = final_data['GDP per capita']
y = final_data['max_infection_rate']
sns.scatterplot(x=x,y=np.log(y))
```
